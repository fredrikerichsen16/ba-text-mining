```python
import pandas as pd
import sqlite3
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import glob
import re
```

```python
def remove_urls(txt):
    return re.sub(r'(https|http)?:\/\/(\w|\.|\/|\?|\=|\&|\%)*\b', '', str(txt), flags=re.MULTILINE)

def replace_usernames(txt):
    return re.sub(r'(^|\s)@(\w+)', ' @user', str(txt), flags=re.MULTILINE)

def preprocess_tweets(tweets):
    return [preprocess_tweet(tweet) for tweet in tweets]

def preprocess_tweet(tweet):
    tweet = remove_urls(tweet)
    tweet = replace_usernames(tweet)
    tweet = tweet.replace('\n', ' ')
    
    return tweet
```

```python
analyzer = SentimentIntensityAnalyzer()

filenames = sorted(glob.glob(r'C:\Users\alvin\Desktop\Bitcoin_tweets\b\*.csv'))

for f in filenames:
    df = pd.read_csv(f)
    df['processedtext'] = df['text'].apply(preprocess_tweet)
    #load the descriptions into textblob
    desc_blob = [TextBlob(txt) for txt in df['processedtext']]
    #add the sentiment metrics to the dataframe
    df['Subj'] = [b.sentiment.subjectivity for b in desc_blob]
    df['compound'] = [analyzer.polarity_scores(v)['compound'] for v in df['processedtext']]
    #df.drop('Unnamed: 0')
    #df.to_csv('{}edited.csv'.format(f), index=False)
df.head(10)
```
