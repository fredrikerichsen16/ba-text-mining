```python
df = pd.read_csv(os.path.abspath('data/Bitcoin_tweets.csv'))
```

```python
df = df[-1000:]
```

```python
df_sorted = df.sort_values(by='user_favourites', ascending=False)
```

```python
df_sorted.head(20)
```

```python
tweetz = list(df_sorted.text)
```

```python
tweetz[347]
```

```python
idxs = [3, 4, 5, 6, 7, 8, 10, 13, 14, 19, 21, 22, 93, 97, 112, 118, 119, 125, 128, 129, 130, 133, 135, 140, 142, 160, 161, 167, 168, 172, 173, 176, 177, 180, 182, 201, 204, 205, 220, 223, 228, 240, 243, 245, 247, 250, 256, 261, 263, 278, 284, 292]
```

```python
"1": {
    "sentiment_label": "",
    "text_of_tweet": "",
    "tweet_url": ""
}
```

```python
import json
```

```python
obj = {}

for idx, i in enumerate(idxs):
    obj[str(idx)] = {
        "sentiment_label": 0,
        "text_of_tweet": tweetz[i],
        "tweet_url": ""
    }

# Turn object into JSON string
json_str = json.dumps(obj)

# Write to file
with open('data/tweets.json', 'w') as f:
    f.write(json_str)
```
